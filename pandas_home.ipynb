```python
import pandas as pd
print(pd.__version__)
```

```python
df = pd.read_csv('student_admission_record_dirty.csv')
df
```

```python
df.info()
df
```

```python
missing_ratio = df.isnull().mean()*100
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
# High School Percentage 결측치를 평균값으로 대체
df['High School Percentage'] = df['High School Percentage'].fillna(df['High School Percentage'].mean())

# 확인
print(df.isnull().sum())
df.head()
```

```python
df['City'] = df['City'].fillna(df['City'].mode()[0])
df
```

```python
df['Admission Status'] = df['Admission Status'].fillna('Unknown')
df
```

```python
print(df.isnull().sum())
df.head()
```
